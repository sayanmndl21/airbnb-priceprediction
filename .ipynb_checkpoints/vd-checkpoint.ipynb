# EDA

### Reduce number of values of some features/ change feature types etc

```python
train_data_rec = train_data.copy()
test_data_rec = test_data.copy()

###### ###### ###### ###### ###### ###### ###### ###### ###### ###### 
###### do same for test data
###### ###### ###### ###### ###### ###### ###### ###### ###### ###### 

# neighbourhood
places, counts = np.unique(train_data_rec['neighbourhood'], return_counts=True)
for i, pl in enumerate(places[counts < 50]):
    train_data_rec.replace({'neighbourhood':{pl:'Other'}}, inplace=True)
    test_data_rec.replace({'neighbourhood':{pl:'Other'}}, inplace=True)
    
# last reviews
tr_t1 = np.array([datetime.strptime(d, '%m/%d/%y') for d in train_data_rec['last_review']])
te_t1 = np.array([datetime.strptime(d, '%m/%d/%y') for d in test_data_rec['last_review']])

tr_t2 = tr_t1 - max(tr_t1)
te_t2 = te_t1 - max(tr_t1)

tr_t3 = -np.array([ti.days for ti in tr_t2])
train_data_rec['last_review'] = tr_t3.astype(float)
te_t3 = -np.array([ti.days for ti in te_t2])
test_data_rec['last_review'] = te_t3.astype(float)

# host since
tr_t1 = np.array([datetime.strptime(d, '%m/%d/%y') for d in train_data_rec['host_since']])
te_t1 = np.array([datetime.strptime(d, '%m/%d/%y') for d in test_data_rec['host_since']])

tr_t2 = tr_t1 - max(tr_t1)
te_t2 = te_t1 - max(tr_t1)

tr_t3 = -np.array([ti.days for ti in tr_t2])
train_data_rec['host_since'] = tr_t3.astype(float)
te_t3 = -np.array([ti.days for ti in te_t2])
test_data_rec['host_since'] = te_t3.astype(float)



```

# Clustering

```python
# one-hot encoding

tr_dat_n = pd.get_dummies(train_data_rec, 
               prefix = ['neighbourhood', 'room_type', 'host_is_superhost', 
                         'bed_type', 'instant_bookable', 'cancellation_policy', 
                         'guest_profile', 'guest_phone'], 
               columns = ['neighbourhood', 'room_type', 'host_is_superhost', 
                          'bed_type', 'instant_bookable', 'cancellation_policy', 
                          'require_guest_profile_picture', 
                          'require_guest_phone_verification'], 
               drop_first=True) 

tr_dat_n.drop(['id', 'is_business_travel_ready'], axis=1, inplace=True)

y = tr_dat_n['price']
tr_dat_n.drop(['price'], axis = 1, inplace = True)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(tr_dat_n)

km = KMeans(n_clusters=8, random_state=0).fit(x)
```

```python
np.unique(km.labels_, return_counts = True)
```

```python
np.unique(y, return_counts = True)
```

```python
sk.metrics.confusion_matrix(
    1 + km.labels_, 
    y)
```

# XGBoost

```python
import xgboost as xgb
```

```python
# one-hot encoding

# merge, encode, split

# merge
ytrain = train_data_rec['price']
Xtrain_xgb = train_data_rec.drop(['price'], axis=1)
ntr = Xtrain_xgb.shape[0]

data_comp = pd.concat([Xtrain_xgb, test_data_rec])

# encode
data_comp = pd.get_dummies(data_comp, 
               prefix = ['neighbourhood', 'room_type', 'host_is_superhost', 
                         'bed_type', 'instant_bookable', 'cancellation_policy', 
                         'guest_profile', 'guest_phone'], 
               columns = ['neighbourhood', 'room_type', 'host_is_superhost', 
                          'bed_type', 'instant_bookable', 'cancellation_policy', 
                          'require_guest_profile_picture', 
                          'require_guest_phone_verification'], 
               drop_first=True) 

# split
xtrain = data_comp.iloc[0:ntr, :]
xtest = data_comp.iloc[ntr:,:]

xtrain.drop(['id', 'is_business_travel_ready'], axis=1, inplace=True)
xtest.drop(['id', 'is_business_travel_ready'], axis=1, inplace=True)


# test
#Xtest_xgb = pd.get_dummies(test_data_rec, 
#               prefix = ['neighbourhood', 'room_type', 'host_is_superhost', 
#                         'bed_type', 'instant_bookable', 'cancellation_policy', 
#                         'guest_profile', 'guest_phone'], 
#               columns = ['neighbourhood', 'room_type', 'host_is_superhost', 
#                          'bed_type', 'instant_bookable', 'cancellation_policy', 
#                          'require_guest_profile_picture', 
#                          'require_guest_phone_verification'], 
#               drop_first=True) 

#ytrain_xgb = Xtrain_xgb['price']
#Xtest_xgb.drop(['id', 'is_business_travel_ready'], axis=1, inplace=True) #,'price'
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(Xtrain_xgb, test_size=0.1)
```

```python
ytr = train['price']
xtr = train.drop(['price'], axis = 1)
yte = test['price']
xte = test.drop(['price'], axis = 1)
```

```python
# training 
xgb1 = xgb.sklearn.XGBClassifier(learning_rate=0.1,
                    n_estimators=400,
                    max_depth=6,
                    min_child_weight=1,
                    gamma=0,
                    subsample=1,
                    colsample_bytree=1,
                    objective='multi:softmax',
                    nthread=-1,
                    num_class=4,
                    seed=96,
                                eval_metric = 'mae')

xgb1.fit(xtr, ytr) 
```

```python
preds = xgb1.predict(xte)
```

```python
sum(abs(preds - yte))
```

```python
sum(preds == yte)/yte.shape[0]
```
